#Function

```python
def multiply_num(*args):
  result = 1
  for num in args:
    result*=num
  return result
  
```

```python
result = multiply_num(2,3,4)
result
```

```python
def greet(name,**kwargs):
  if 'age' in kwargs:
    return f"Hello, {name}! You are {kwargs['age']} years old."
  else:
    return f"Hello, {name}!"
```

```python
result1 = greet("Alice")
result2 = greet("Bob",age=25)
```

```python
print(result1)
print(result2)
```

```python
def apply_operation(a,b,operation):
  return operation(a,b)
```

```python
def multiply(a,b):
  return a*b
```
