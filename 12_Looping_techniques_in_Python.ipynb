#Enumerate() in python

```python
days = ["Sunday","Monday","Tuesday","Wednesday"]
```

```python
enum_days = enumerate(days)
print(enum_days)
```

```python
list(enum_days)
```

```python
days = ["Sunday","Monday","Tuesday","Wednesday"]
enum_days = enumerate(days,5)
list(enum_days)
```

```python
days = ["Sunday","Monday","Tuesday","Wednesday"]
enum_days = enumerate(days,5)
for day in enum_days:
  print(day)
```

```python
days = ["Sunday","Monday","Tuesday","Wednesday"]
for count,day in enumerate(days):
  print(count," : ",day)
```

```python
days = ["Sunday","Monday","Tuesday","Wednesday"]
for count,day in enumerate(days,1):
  print(count," : ",day)
```

#Zip

```python
characters = ["Mickey","Donald","Goofy","Simba"]
fav = ["Cheese","Jazz","Cats","Fish"]
zip_me = zip(characters,fav)
```

```python
print(zip_me)
```

```python
tuple(zip_me)
```

```python
characters = ["Mickey","Donald","Goofy","Simba"]
fav = ["Cheese","Jazz","Cats","Fish"]
numbers = [1,2,3,5]
zip_me = zip(characters,fav,numbers)
tuple(zip_me)
```

```python
characters = ["Mickey","Donald","Goofy","Simba"]
fav = ["Cheese","Jazz","Fish"]
numbers = [1,2,3,5]
zip_me = zip(characters,fav,numbers)
tuple(zip_me)
```

```python
characters = ["Mickey","Donald","Goofy","Simba"]
fav = ["Cheese","Jazz","Fish"]
numbers = [1,2,3,5]
zip_me = zip(characters,fav,numbers)
list(zip_me)
```

#Unzip

```python
characters = ["Mickey","Donald","Goofy","Simba"]
fav = ["Cheese","Jazz","Cats","Fish"]

zip_me = zip(characters,fav)
```

```python
c,z= zip(*zip_me)
print(c)
print(z)
```

#Looping through multiple lists at once

```python
characters = ["Mickey","Donald","Goofy","Simba"]
favs = ["Cheese","Jazz","Cats","Fish"]
for character,fav in zip(characters,favs):
  print(character," : ",fav)
```

#sorted()

```python
a = (1,11,2)
x = sorted(a)
print(x)
```

```python
characters = ["Mickey","Donald","Goofy","Simba"]
for character in sorted(characters):
  print(character)
```

```python
novel = {"Name":"Karnali Blues",
         "Writer":"Buddhi Sagar",
         "Rating":4.2}
for key in sorted(novel):
  print(key)
```

```python
novel = {"Name":"Karnali Blues",
         "Writer":"Buddhi Sagar",
         "Rating":4.2}
for item in sorted(novel.items()):
  print(item)
```

#Reversed

```python
novel = {"Name":"Karnali Blues",
         "Writer":"Buddhi Sagar",
         "Rating":4.2}
for item in reversed(novel.items()):
  print(item)
```

```python
characters = ["Mickey","Donald","Goofy","Simba"]
for character in reversed(characters):
  print(character)
```

#Creating dictionary with zipped keys and values

```python
characters = ["Mickey","Donald","Goofy","Simba"]
favs = ["Cheese","Jazz","Cats","Fish"]
characters_dict = dict(zip(characters,favs))
characters_dict
```
