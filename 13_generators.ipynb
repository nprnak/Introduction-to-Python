```python
def create_emails(person_list):
  emails = []
  for name in person_list:
    email = name+"@knight.com"
    emails.append(email)

  return emails
```

```python
def create_emails_gen(person_list):
  for name in person_list:
    email = name+"@knight.com"
    yield email
```

```python
person_list = ['Pravin',"Thaneswor","Roshan","Rohan"]
```

```python
emails = create_emails(person_list)
print(emails)
```

```python
emails = create_emails_gen(person_list)
print(emails)
```

```python
emails = create_emails_gen(person_list)
print(next(emails))
```
