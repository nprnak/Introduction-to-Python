#Cross tabulation in pandas

```python
path = "/content/jpt.csv"
```

```python
df = pd.read_csv(path)
df.tail()
```

```python
df.columns
```

```python
len(df.columns)
```

```python
pd.crosstab(index=df.Country,columns=df.Gender)
```

```python
pd.crosstab(index=df['Country'],columns=df['Gender'])
```

```python
pd.crosstab(df['Country'],df['Gender'])
```

```python
pd.crosstab([df['Country'],df['City']],df['Gender'])
```

```python
pd.crosstab(df['Country'],[df['Gender'],df['Movie']])
```

```python
pd.crosstab([df['Country'],df['City']],[df['Gender'],df['Movie']])
```

#Selection with .loc and .iloc

```python
pd.crosstab(df['Country'],df['Gender']).loc['Germany']
```

```python
pd.crosstab([df['Country'],df['City']],df['Gender']).loc['Germany']
```

```python
pd.crosstab([df['Country'],df['City']],df['Gender']).loc['Germany','Berlin']
```

```python
pd.crosstab([df['Country'],df['City']],df['Gender']).loc['Berlin']
```

```python
pd.crosstab([df['Country'],df['City']],df['Gender']).loc[['Germany','Peru'],['F']]
```

```python
pd.crosstab([df['Country'],df['City']],[df['Gender'],df['Cheese']]).loc[[('Germany','Berlin'),("Peru","Tacna")]]
```

```python
list_of_selection = [('Germany','Berlin'),("Peru","Tacna")]
pd.crosstab([df['Country'],df['City']],[df['Gender'],df['Cheese']]).loc[list_of_selection]
```

```python
pd.crosstab([df['Country'],df['City']],[df['Gender'],df['Cheese']])
```

```python
pd.crosstab([df['Country'],df['City']],[df['Gender'],df['Cheese']]).iloc[:3]
```

#using basic function with crosstab

```python
pd.crosstab([df.City],[df.Gender]).loc[['Kabul','Lima',"Marka"]].max()
```

```python
#margins are for total
pd.crosstab([df.City],[df.Gender],margins=True)
```

```python
#normalizing results
pd.crosstab([df.Country],[df.Gender],normalize=True)
```

```python
pd.crosstab([df.Country],[df.Gender],normalize="index")
```

```python
pd.crosstab([df['Country'],df['City']],[df.Gender],normalize="columns")
```
