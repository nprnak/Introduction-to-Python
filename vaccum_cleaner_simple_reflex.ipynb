## 1. Theory

### 1.1 What is a Simple Reflex Agent?

A **Simple Reflex Agent** is the most basic type of intelligent agent in Artificial Intelligence that selects actions based solely on the **current percept** (sensor input at that moment). It operates using **condition-action rules** (if-then statements) and has **no memory** of past states or history.

**Key Characteristics:**
- Responds only to current environmental stimuli
- Uses predefined rule sets (condition → action mapping)
- Does not maintain any internal state or memory
- Works best in fully observable environments
- Extremely fast response time

**Architecture:**

Percept (Current) → Condition-Action Rules → Action


### 1.2 Vacuum Cleaner World Problem

The Vacuum Cleaner World is a classic AI problem used to demonstrate agent behavior.

**Environment Specifications:**
- **Two rooms**: Room A and Room B
- **States**: Each room can be either "Clean" or "Dirty"
- **Agent Location**: The vacuum agent can be in either Room A or Room B
- **Fully Observable**: Agent always knows its location and room status

**Possible Actions:**
| Action | Description |
|--------|-------------|
| `Suck` | Clean the current room (removes dirt) |
| `Move to A` | Navigate from current room to Room A |
| `Move to B` | Navigate from current room to Room B |
| `NoOp` | No operation (idle/do nothing) |

### 1.3 Working Principle of the Agent

The agent follows a **pre-programmed rule set** that maps each possible percept to an appropriate action:

**Rule Table:**

| Condition (Percept) | Action | Rule Number |
|--------------------|--------|-------------|
| Current room is Dirty | Suck | R1 |
| Current room is Clean AND located in Room A | Move to B | R2 |
| Current room is Clean AND located in Room B | Move to A | R3 |
| Any other case (fallback) | NoOp | R4 (Default) |

**Explanation of Rules:**
- **R1**: If dirt is detected, clean immediately (highest priority)
- **R2**: If Room A is clean, go check Room B
- **R3**: If Room B is clean, go check Room A  
- **R4**: Default fallback when no condition matches


### Step-by-Step Execution Algorithm
Step 1: Start
Step 2: Read current percept (location, status) from sensors
Step 3: Evaluate status condition
Step 3.1: IF status equals "Dirty" THEN
Step 3.1.1: Set action = "Suck"
Step 3.1.2: Goto Step 7
Step 4: Evaluate location condition
Step 4.1: IF location equals "A" THEN
Step 4.1.1: Set action = "Move to B"
Step 4.1.2: Goto Step 7
Step 5: Evaluate second location condition
Step 5.1: IF location equals "B" THEN
Step 5.1.1: Set action = "Move to A"
Step 5.1.2: Goto Step 7
Step 6: If no conditions match
Step 6.1: Set action = "NoOp"
Step 7: Return/Output the action
Step 8: End

```python
def simple_reflex_agent(location, status):
    # location: 'A' or 'B', status: 'Dirty' or 'Clean'
    if status == "Dirty":
        return "Suck"
    elif location == "A":
        return "Move to B"
    elif location == "B":
        return "Move to A"
    else:
        return "NoOp"


# Percept Sequences to test
percepts = [
    ("A", "Dirty"),
    ("A", "Clean"),
    ("B", "Dirty"),
    ("B", "Clean")
]

print("********** Simple Reflex Agent **********\n")
print("Program by: [Your Name]")
print("Roll no: [Your Roll No]\n")

for percept in percepts:
    location, status = percept
    action = simple_reflex_agent(location, status)
    print(f"Percept: Location={location}, Status={status} → Action: {action}")
```
