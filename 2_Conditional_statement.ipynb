#And Or

```python
marks_math = int(input("Enter your marks in math: "))
marks_english = int(input("Enter your marks in English: "))

if marks_math > 60 and marks_english > 60:
  print("You are a good student")
```

```python
marks_math = int(input("Enter your marks in math: "))
marks_english = int(input("Enter your marks in English: "))

if marks_math > 60 or marks_english > 60:
  print("You are a good student")
```
