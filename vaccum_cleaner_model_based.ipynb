## 1. Theory

### 1.1 What is a Model-Based Agent?

A **Model-Based Agent** is an intelligent agent that maintains an **internal state** (memory) to keep track of the world's history and unobserved aspects. Unlike simple reflex agents, it has knowledge of how the world evolves and uses this **world model** to make decisions.

**Key Characteristics:**
- Maintains an internal representation of the environment
- Updates its model based on current percepts
- Can handle partially observable environments
- Remembers past states and actions
- Uses both current percept and internal state for decision making

**Architecture:**

Percept (Current) + Internal State → Update Model → Condition-Action Rules → Action


### 1.2 Vacuum Cleaner World Problem (Model-Based Approach)

**Environment Specifications:**
- **Two rooms**: Room A and Room B
- **States**: Each room can be either "Clean" or "Dirty"
- **Agent Location**: The agent can be in either Room A or Room B
- **Partially Observable**: Agent may not always know the status of both rooms
- **Internal Model**: Agent maintains memory of known room statuses

**World Model (Internal State):**

world_model = {
"A": "Clean" or "Dirty",
"B": "Clean" or "Dirty"
}


**Possible Actions:**
| Action | Description |
|--------|-------------|
| `Go to A and Suck` | Navigate to Room A and clean it |
| `Go to B and Suck` | Navigate to Room B and clean it |
| `Do Nothing, all are clean` | Idle state when both rooms are clean |

### 1.3 Working Principle

The model-based agent maintains an internal **world model** that gets updated with each percept. It then uses this model to decide which room needs cleaning.

**Step-by-Step Process:**

1. **Perceive**: Agent receives current percept (location, status)
2. **Update Model**: Updates `world_model[location]` with the new status
3. **Query Model**: Checks the internal memory for dirty rooms
4. **Act**: Takes action based on the complete model

**Rule Table:**

| Condition (World Model) | Action |
|------------------------|--------|
| Room A is Dirty | Go to A and Suck |
| Room B is Dirty | Go to B and Suck |
| Both Rooms are Clean | Do Nothing, all are clean |

### Step-by-Step Execution Algorithm
Step 1: Start
Step 2: Initialize world_model = {"A": "Clean", "B": "Clean"}
Step 3: Read current percept (location, status)
Step 4: Update internal memory
Step 4.1: Set world_model[location] = status
Step 5: Query the updated model
Step 5.1: IF world_model["A"] == "Dirty" THEN
Step 5.1.1: Set action = "Go to A and Suck"
Step 5.1.2: Goto Step 6

Step 5.2: ELSE IF world_model["B"] == "Dirty" THEN
Step 5.2.1: Set action = "Go to B and Suck"
Step 5.2.2: Goto Step 6

Step 5.3: ELSE
Step 5.3.1: Set action = "Do Nothing, all are clean"
Step 6: Return/Output the action
Step 7: End (but internal state persists for next cycle) 

```python
# Lab 1: Model-Based Agent for Vacuum Cleaner World
# Program by: [Your Name]
# Roll no: [Your Roll No]

# Agent's internal memory (model of the world)
world_model = {"A": "Clean", "B": "Clean"}

def model_based_agent(location, status):
    # Update the agent's memory
    world_model[location] = status
    
    # Condition-Action Rules using memory
    if world_model["A"] == "Dirty":
        return "Go to A and Suck"
    elif world_model["B"] == "Dirty":
        return "Go to B and Suck"
    else:
        return "Do Nothing, all are clean"

print("********** Model-Based Agent **********\n")
print("Program by: [Your Name]")
print("Roll no: [Your Roll No]\n")

print("Initially, the state model contains:", world_model)

print("\nThe percept sequence is (A, Dirty)")
action = model_based_agent("A", "Dirty")
print("Action:", action)
print("The state model contains:", world_model)

print("\nThe percept sequence is (A, Clean)")
action = model_based_agent("A", "Clean")
print("Action:", action)
print("The state model contains:", world_model)

print("\nThe percept sequence is (B, Dirty)")
action = model_based_agent("B", "Dirty")
print("Action:", action)
print("The state model contains:", world_model)

print("\nThe percept sequence is (B, Clean)")
action = model_based_agent("B", "Clean")
print("Action:", action)
print("The state model contains:", world_model)
```
