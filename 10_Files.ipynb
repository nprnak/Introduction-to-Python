#Writing to an existing file

```python
my_file = open("neptune.txt",'a')
```

```python
my_file.write("\n Adding new lines to file")
```

```python
my_file.close()
```

```python
my_file = open("neptune.txt")
my_file.read()
```

```python
my_file.close()
```

```python
#overwriting content of neptune.txt
my_file = open("neptune.txt","w")
my_file.write("\n Neptune has been replaced by Pluto")
my_file.close()
```

```python
my_file = open("neptune.txt")
my_file.read()
```
