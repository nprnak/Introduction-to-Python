#using a functio with a while loop

```python
def musician(artist,instrument):
  output = artist + ' plays '+instrument
  return output
```

```python
while True:
  name = input("What is your name")
  instrument =input("What do tou play?")
  formatted_name = musician(name,instrument)
  print("\n Hello "+formatted_name+"!" )
  response = input("Do you want to continue?")

  if (response == 'no'):
    break
```

#Passing a list

```python
def greet_user(names):
  for name in names:
    msg = "Hello "+name.title()
    print(msg)
```

```python
usernames = ['Mickey',"Donald","Winne",'Lui']
greet_user(usernames)
```

#Returning a list

```python
def list_of_staff():
  staff_list = []
  while True:
    staff = input("Enter name of a staff")
    staff_list.append(staff)
    response = input("Enter 'Q' to exit: ")
    if response == 'Q':
      break
  return staff_list
```

```python
staffs = list_of_staff()
```

```python
staffs
```

#Passing an arbitary number of arguments

```python
def staff_list(*staff_name):
  for staff in staff_name:
    print(staff)
```

```python
staff_list('Mickey')
```

```python
staff_list('Mickey',"goofy","Donald")
```

```python
def student(name,*skills):
  print(name+ " has following skills")
  for skill in skills:
    print(skill)
```

```python
student("Durga","Running","Playing")
```

#Using arbitary keyword arguments

```python
def build_profile(first,last,**user_info):
  profile = {}
  profile['first_name']=first
  profile['last_name'] = last
  for key,value in user_info.items():
    profile[key]=value
  return profile
```

```python
person1 = build_profile("Mickey","Mouse",age=55,height=150)
person2 = build_profile("Donald","Duck",age=55,height=30,hobby=['chess','ludo','cards'])
```

```python
person1
```
