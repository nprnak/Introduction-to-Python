#Unpacking Tuple

```python
numbers = (2,3,5)
```

```python
num,num2,num3 = numbers
print(num)
print(num2)
print(num3)
```

```python
numbers = (1,2,5,4,7,8,9,6,3,1)
```

```python
num1,num2,*num3 = numbers
print(num1)
print(num2)
print(num3)
```

```python
num1,*num2,num3 = numbers
print(num1)
print(num2)
print(num3)
```

```python
num1,num2,num3 = numbers
print(num1)
print(num2)
print(num3)
```

```python
*num1, = numbers
print(num1)
```

#Loops with tuple

```python
coelho = ('Brida','Alchemist','Veronica','Zahir')
```

```python
for books in coelho:
  print(books)
```

```python
count =0
for book in coelho:
  if 'r' in book:
    count += 1
print(count)
```

```python
i=0
while i<len(coelho):
  print(coelho[i])
  i +=1
```
