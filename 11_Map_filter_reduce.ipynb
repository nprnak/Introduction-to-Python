#Map

```python
my_pets = ['cat','brown','purple','pluto']
```

```python
my_pets_u = []
for pet in my_pets:
  petU = pet.upper()
  my_pets_u.append(petU)

print(my_pets_u)
```

```python
map_pets = list(map(str.upper,my_pets))
print(map_pets)
```

```python
round(5.8292996,3)
```

```python
circle_areas = [3.6541,6.489955,3.2147,55.2489651,2.369,1.2540,12.354788521]
result = list(map(round,circle_areas))
print(result)
```

```python
circle_areas = [3.6541,6.489955,3.2147,55.2489651,2.369,1.2540,12.354788521]
result = list(map(round,circle_areas,range(1,7)))
print(result)
```

```python
numbers = [1,2,3,4,5,6,7]
powers = [1,2,5,8,9,6,3]
result = list(map(lambda x,y: x**y,numbers,powers))
print(result)
```

```python
my_string = ['a','b','c','d','e']
my_numbers = [1,2,3,4,5]
results = list(map(lambda x,y:(x,y),my_string,my_numbers))
print(results)
```

#Filter

```python
scores = [66,99,88,44,55,22,66,33]
def is_A_student(score):
  return score>75
```

```python
is_A_student(15)
```

```python
over_75 = list(filter(is_A_student,scores))
```

```python
over_75
```

```python
list(filter(lambda x:x>75,scores))
```

#Reduce

```python
from functools import reduce
```

```python
numbers = [1,2,3,4,5]
def custom_sum(first,second):
  return first+second
```

```python
result = reduce(custom_sum,numbers)
print(result)
```

```python
nums = [1,2]
reduce(lambda x,y:x+y**y,nums,5)
```

```python
matrix = [[1,2,3],[2,4,5],[9,8,7]]
reduce(lambda x,y:x+y,matrix)
```

#Map and filter on same function

```python
countries = ["Afganisthan","China","Egypt","Libiya","Austria","Spain","Iran"]
```

```python
def A_countries(country):
  return country[0] == 'A'
```

```python
A_countries("Algeria")
```

```python
countries_A = list(map(A_countries,countries))
print(countries_A)
```

```python
countries_A = list(filter(A_countries,countries))
print(countries_A)
```
