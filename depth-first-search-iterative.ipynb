<h2 class="section-title">Part 1: Background</h2>
<div class="card">
    <h3><span class="badge">1</span> What is Depth-First Search</h3>
    <div class="theory-box">
        <p><strong>Definition:</strong> Depth-First Search (DFS) is an uninformed search algorithm that explores a graph by going as deep as possible along each branch before backtracking. It starts at the root node and explores one branch completely before moving to the next branch.</p>
    </div>
    <div class="theory-box">
        <p><strong>Simple Analogy:</strong></p>
        <p>Imagine you are searching for a book in a library. DFS is like going to one shelf, checking every book on it, then going to the next shelf, and so on. You explore one path completely before moving to another path.</p>
    </div>
    <div class="theory-box">
        <p><strong>Key Characteristics:</strong></p>
        <ul>
            <li>Uses a Stack (LIFO) data structure &ndash; last in, first out</li>
            <li>Does NOT guarantee the shortest path</li>
            <li>Not complete &ndash; may go down an infinite path</li>
            <li>Not optimal &ndash; may find a solution that is not the shortest</li>
            <li>Memory efficient &ndash; stores only the current path</li>
        </ul>
    </div>
    <div class="tree"><strong>Example Tree (DFS traversal order - left first):</strong>
        <pre>        A (Level 0)
       / \
      B   C (Level 1)
     / \   \
    D   E   F (Level 2)
   /   / \   \
  G   H   I   J (Level 3)

DFS order (left-first): A, B, D, G, E, H, I, C, F, J</pre>
    </div>
</div>
<div class="card">
    <h3><span class="badge">2</span> DFS Algorithm</h3>
    <div class="step-box">
        <p><strong>Algorithm Steps (Iterative using Stack):</strong></p>
        <ol>
            <li>Start with the initial node and push it onto a stack</li>
            <li>Also maintain a visited list to avoid revisiting nodes</li>
            <li>While the stack is not empty:
                <ol type="a">
                    <li>Pop the top node from the stack</li>
                    <li>If the node has not been visited, process (print) it and mark as visited</li>
                    <li>Push all unvisited neighbors onto the stack</li>
                </ol>
            </li>
        </ol>
    </div>
    <div class="code-box">
        <pre># DFS Algorithm Pseudocode (Iterative using Stack)
visited = []
stack = []

def dfs(visited, graph, node):
    stack.append(node)
    visited.append(node)

    while stack:
        m = stack.pop()
        print(m, end=" ")

        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                stack.append(neighbour)
</pre>
    </div>
    <div class="step-box">
        <p><strong>Algorithm Steps (Recursive):</strong></p>
        <ol>
            <li>Mark the current node as visited and print it</li>
            <li>For each neighbor of the current node:
                <ol type="a">
                    <li>If the neighbor has not been visited, recursively call DFS on it</li>
                </ol>
            </li>
        </ol>
    </div>
    <div class="code-box">
        <pre># DFS Algorithm Pseudocode (Recursive)
visited = []

def dfs(visited, graph, node):
    if node not in visited:
        visited.append(node)
        print(node, end=" ")
        
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)
</pre>
    </div>

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': ['G', 'H'],
    'E': ['I'],
    'F': ['J'],
    'G': [],
    'H': [],
    'I': [],
    'J': []
}
```

```python
visited = []  # List for visited nodes
stack = []    # Initialize a stack
```

```python
def dfs_iterative(visited, graph, node):
    stack.append(node)
    visited.append(node)

    while stack:
        m = stack.pop()
        print(m, end=' ')

        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                stack.append(neighbour)
    return visited, stack
```

```python
print("=" * 50)
print("DEPTH-FIRST SEARCH TRAVERSAL (ITERATIVE)")
print("=" * 50)
print("DFS Order: ", end='')

visited, stack = dfs_iterative(visited, graph, 'A')

print()  # Adds a new line right after the BFS traversal characters print out
print()  # Adds an extra blank line for visual spacing
print("Visited Nodes: ", visited)
print("Stack: ", stack)
print()
print("=" * 50)
print("Program by: Pranaya Nakarmi")
print("Roll No: 1")
```
