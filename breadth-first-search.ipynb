<h2 class="section-title">Part 1: Background</h2>
<div class="card">
    <h3><span class="badge">1</span> What is Breadth-First Search</h3>
    <div class="theory-box">
        <p><strong>Definition:</strong> Breadth-First Search (BFS) is an uninformed search algorithm that explores a graph level by level. It starts at the root node and visits all nodes at the current depth before moving to nodes at the next depth level.</p>
    </div>
    <div class="theory-box">
        <p><strong>Simple Analogy:</strong></p>
        <p>Imagine you are searching for a book in a library with multiple floors. BFS is like checking every shelf on the ground floor first, then every shelf on the first floor, then the second floor, and so on. You will find the closest book first.</p>
    </div>
    <div class="theory-box">
        <p><strong>Key Characteristics:</strong></p>
        <ul>
            <li>Uses a Queue (FIFO) data structure &ndash; first in, first out</li>
            <li>Finds the shortest path in terms of number of edges when all edges have equal cost</li>
            <li>Complete &ndash; will find a solution if one exists</li>
            <li>Optimal &ndash; finds the shallowest goal (fewest steps)</li>
            <li>Memory intensive &ndash; stores all nodes at the current level</li>
        </ul>
    </div>
    <div class="tree"><strong>Example Tree (BFS traversal order):</strong>
        <pre>        A (Level 0)
       / \
      B   C (Level 1)
     / \   \
    D   E   F (Level 2)
   /   / \   \
  G   H   I   J (Level 3)

BFS order: A, B, C, D, E, F, G, H, I, J</pre>
    </div>
</div>
<div class="card">
    <h3><span class="badge">2</span> BFS Algorithm</h3>
    <div class="step-box">
        <p><strong>Algorithm Steps:</strong></p>
        <ol>
            <li>Start with the initial node and add it to a queue</li>
            <li>Also maintain a visited list to avoid revisiting nodes</li>
            <li>While the queue is not empty:
                <ol type="a">
                    <li>Remove the front node from the queue</li>
                    <li>Print or process the current node</li>
                    <li>Generate all its neighbors (child nodes)</li>
                    <li>For each neighbor that has NOT been visited, add it to the visited list and to the back of the queue</li>
                </ol>
            </li>
        </ol>
    </div>
    <div class="code-box">
        <pre># BFS Algorithm Pseudocode
visited = []
queue = []

def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)

    while queue:
        m = queue.pop(0)
        print(m, end=" ")

        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)</pre>
    
</div>

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': ['G', 'H'],
    'E': ['I'],
    'F': [],
    'G': [],
    'H': [],
    'I': []
}

visited = []  # List for visited nodes
queue = []    # Initialize a stack

def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)

    while queue:
        m = queue.pop(0)
        print(m, end=' ')

        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)
                # print("Visited Nodes: ", visited)
                # print("Queue: ", queue)
    return visited, queue
print("=" * 50)
print("BREADTH-FIRST SEARCH TRAVERSAL")
print("=" * 50)
print("BFS Order: ", end='')

visited, queue = bfs(visited, graph, 'A')
print()  # Adds a new line right after the BFS traversal characters print out
print()  # Adds an extra blank line for visual spacing

print("Visited Nodes: ", visited)
print("Queue: ", queue)
print("=" * 50)
print("Program by: Pranaya Nakarmi")
print("Roll No: 1")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from collections import deque

# ── Graph definition ──────────────────────────────────────────────
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': ['G', 'H'],
    'E': ['I'],
    'F': [],
    'G': [],
    'H': [],
    'I': []
}

# ── BFS ───────────────────────────────────────────────────────────
def bfs(graph, start):
    visited = []
    queue = deque([start])
    seen = {start}
    while queue:
        node = queue.popleft()
        visited.append(node)
        for neighbour in graph[node]:
            if neighbour not in seen:
                seen.add(neighbour)
                queue.append(neighbour)
    return visited

bfs_order = bfs(graph, 'A')

# ── Build directed graph ──────────────────────────────────────────
G = nx.DiGraph()
for node, neighbours in graph.items():
    G.add_node(node)
    for nb in neighbours:
        G.add_edge(node, nb)

# ── Hierarchical layout (top-down by BFS level) ───────────────────
levels = {
    0: ['A'],
    1: ['B', 'C'],
    2: ['D', 'E', 'F'],
    3: ['G', 'H', 'I']
}

pos = {}
for level, nodes in levels.items():
    y = -level * 1.8
    xs = [i - (len(nodes) - 1) / 2 for i in range(len(nodes))]
    for node, x in zip(nodes, xs):
        pos[node] = (x * 2.2, y)

# ── Colours by BFS visit order ────────────────────────────────────
cmap = plt.cm.YlGn
node_colors = [cmap(0.3 + 0.6 * (bfs_order.index(n) / (len(bfs_order) - 1)))
               for n in G.nodes()]

# ── Plot ──────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 7))
fig.patch.set_facecolor('#1e1e2e')
ax.set_facecolor('#1e1e2e')

# Edges
nx.draw_networkx_edges(
    G, pos, ax=ax,
    edge_color='#aaaacc',
    arrows=True,
    arrowstyle='-|>',
    arrowsize=20,
    width=1.8,
    connectionstyle='arc3,rad=0.05',
    min_source_margin=22,
    min_target_margin=22,
)

# Nodes
nx.draw_networkx_nodes(
    G, pos, ax=ax,
    node_color=node_colors,
    node_size=900,
    linewidths=2,
    edgecolors='white',
)

# Labels with BFS order number
labels = {n: f"{n}\n#{bfs_order.index(n)+1}" for n in G.nodes()}
nx.draw_networkx_labels(G, pos, labels=labels, ax=ax,
                        font_color='#1a1a2e', font_size=10, font_weight='bold')

# ── Legend ────────────────────────────────────────────────────────
legend_handles = [
    mpatches.Patch(color=cmap(0.3),  label='Visited first'),
    mpatches.Patch(color=cmap(0.65), label='Visited mid'),
    mpatches.Patch(color=cmap(0.9),  label='Visited last'),
]
ax.legend(handles=legend_handles, loc='lower right',
          facecolor='#2a2a3e', edgecolor='#555', labelcolor='white', fontsize=9)

# ── Level annotations ─────────────────────────────────────────────
for level, nodes in levels.items():
    y = -level * 1.8
    ax.text(-5.5, y, f"Level {level}", color='#888', fontsize=9,
            va='center', style='italic')

# ── BFS order strip at bottom ─────────────────────────────────────
order_str = "BFS order:  " + "  →  ".join(bfs_order)
ax.text(0, -6.4, order_str, color='#ccccdd', fontsize=10,
        ha='center', va='center',
        bbox=dict(boxstyle='round,pad=0.4', facecolor='#2a2a3e', edgecolor='#555'))

# ── Finishing touches ─────────────────────────────────────────────
ax.set_title("Breadth-First Search — Graph Visualization", color='white',
             fontsize=14, fontweight='bold', pad=16)
ax.axis('off')
plt.tight_layout()
plt.savefig('bfs_graph.png', dpi=150, bbox_inches='tight',
            facecolor=fig.get_facecolor())
plt.show()
print("=" * 50)
print("BFS Order:", " → ".join(bfs_order))
print("=" * 50)
print("Program by: Pranaya Nakarmi")
print("Roll No: 1")
```
