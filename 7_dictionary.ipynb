#Looping through a dictionary

```python
cellphone= {'name':'r21',
            'Price':1256,
            'Bluetooth':True,
            "Dualsim":False,
            "Ram":4,
            "Cores":6,
            "colors":('Black',"Brown","Red")}
```

```python
cellphone
```

```python
for x in cellphone:
  print(x)
```

```python
for x in cellphone.keys():
  print(x)
```

```python
for x in cellphone:
  print(cellphone[x])
```

```python
for x in cellphone.values():
  print(x)
```

```python
for key,value in cellphone.items():
  print(key," : ",value)
```

#Dictionaries in Dictionary

```python
shop = {
    'cellphoneOne':{
        "name":"R22",
        "Price":546,
        "Ram":4,
        "Colors":('Red','Black','Brown')
    },
    'cellphoneTwo':{
        "name":"R55",
        "Price":5466,
        "Ram":6,
        "Colors":('Yellow','Red','Black','Brown')
    },
    'cellphoneThree':{
        "name":"C78",
        "Price":1546,
        "Ram":8,
        "Colors":('Black','Brown')
    }

}
```

```python
shop
```

```python
len(shop)
```

```python
for x in shop:
  print(x)
```

```python
for key,value in shop.items():
  print(key," : ",value)
  print("...............\n")
```

```python
for key,value in shop.items():
  print(key)
  if type(value) == dict:
    for key2,value2 in value.items():
      print(key2," : ",value2)
  print("..........................")
  print("\n")
```

```python
shirt1 = {"logo":False,'Color':('red','White'),'Price':2000,'Slogan':"Infinity and beyond"}
shirt2 = {"logo":True,'Color':('red','White','Pink'),'Price':1000,'Slogan':"Random walk"}
shirt3 = {"logo":"Bird",'Color':('red','White','Purple'),'Price':500,'Slogan':False}
```

```python
shirts = {"shirt1":shirt1,"shirt2":shirt2,"shirt3":shirt3}
```

```python
shirts
```

```python
for key,value in shirts.items():
  print(key)
  if type(value) == dict:
    for key2,value2 in value.items():
      print(key2," : ",value2)
  print("..........................")
  print("\n")
```
