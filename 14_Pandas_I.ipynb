#DataFrame in Pandas

```python
Math = pd.Series([10,20,30,40],
                 index= ["Saurav","Prakash","Sandesh","Bibek"])
Compiler = pd.Series([20,22,32,46],
                 index= ["Saurav","Prakash","Sandesh","Bibek"])
SE = pd.Series([60,88,99,15],
                 index= ["Saurav","Prakash","Sandesh","Bibek"])
E_gov = pd.Series([11,20,50,40,69],
                 index= ["Amrit","Saurav","Prakash","Sandesh","Bibek"])
```

```python
marks = {"Math":Math,"Compiler":Compiler,"SE":SE,"E Gov":E_gov}
```

```python
df = pd.DataFrame(marks)
```

```python
df
```

```python
df2 = pd.DataFrame(marks,index = ["Bibek","Sandesh","Prakash"])
print(df2)
```

```python
df3 = pd.DataFrame(marks,index = ["Bibek","Sandesh","Prakash"],
                   columns=['Math',"Compiler"])
print(df3)
```

```python
df3 = pd.DataFrame(marks,index = ["Bibek","Sandesh","Prakash","Dipesh"],
                   columns=['Math',"Compiler"])
print(df3)
```

```python
df3 = pd.DataFrame(marks,index = ["Bibek","Sandesh","Prakash","Dipesh"],
                   columns=['Math',"Compiler","TOC"])
print(df3)
```

```python
dict_a = {"Math":[10,22,33],"Compiler":[88,99,66],"SE":[88,44,55]}
df = pd.DataFrame(dict_a)
df
```

```python
dict_a = {"Math":[10,22,33],"Compiler":[88,99,66],"SE":[88,44,55]}
df = pd.DataFrame(dict_a,index=["Saurav","Prakash","Sandesh"])
df
```

#Selecting, adding and deleting data in datafrme

```python
expense= {'Food':[55,66,88,99,55],
          "Travel":[77,88,99,12,63],
          "Rent":[77,54,12,63,96]}
```

```python
df = pd.DataFrame(expense)
df
```

```python
df['Total'] = df['Food']+df['Travel']+df['Rent']
df
```

```python
df["Flag"] =df["Total"] > 200
df
```

```python
df["Random"] = "ABC"
df
```

```python
random = df.pop('Random')
df
```

```python
random
```

```python
del df['Rent']
df
```

```python
df.insert(0,'Education',[5,88,99,66,33])
df
```

```python
df.insert(2,"Income",[88,152,123,456,748])
df
```

```python
df = df.assign(Net = lambda x : x['Total']-x['Income'])
df
```
